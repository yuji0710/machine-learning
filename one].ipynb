```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load your dataset
df = pd.read_csv('one2.csv')

# Train the model
reg = LinearRegression()
reg.fit(df[['index']], df['age'])  # Ensure 'index' is a DataFrame
# Plot original data
plt.scatter(df['index'], df['age'], color='red', marker='+', label="Actual Data")



# Generate predictions for the full range
x_values = np.array(df['index']).reshape(-1, 1)  # Convert to 2D array
y_pred = reg.predict(x_values)  # Predict for all x values

# Plot the regression line
plt.plot(df['index'], y_pred, color='blue', linewidth=2, label="Regression Line")



# Predict a specific value (e.g., index=121)
x_pred = np.array([[121]])  # Must be 2D
y_pred_single = reg.predict(x_pred)
# Mark the predicted value
plt.scatter(121, y_pred_single, color='green', marker='o', label="Predicted Point")

# Labels and title
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Linear Regression Prediction')
plt.legend()

# Show the plot
plt.show()
```
