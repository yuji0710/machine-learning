```python
df=pd.DataFrame(wine.data,columns=wine.feature_names)
df
```

```python
df['target']=wine.target
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(wine.data,wine.target,test_size=0.3,random_state=100)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
mn=MultinomialNB()
mn.fit(x_train,y_train)
mn.score(x_test,y_test)
```

```python
y_pred=mn.predict(x_test)
```

```python
x1=x_test[:,0]
x2=x_test[:,9]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
s=plt.scatter(x1,x2,c=y_pred,cmap='viridis',edgecolor='k')
plt.xlabel(wine.feature_names[0])
plt.ylabel(wine.feature_names[9])
plt.title('Navie Bayes Prediction (Test set)')
plt.colorbar(s,label='Predict Wine Class')
plt.show()
```
